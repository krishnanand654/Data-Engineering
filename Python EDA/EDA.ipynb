#### 7. Handling missing values
- mean if data is continous
- median if data is continous or discrete
- mode if data is categorical
- avoided is missing data more than 5%
- more than 20% remove column

````python
import numpy as np
````

````python
#replace '? ' by 'nan' and update DataFrame
raw_data.replace('?', np.nan, inplace = True)
````

````python
# returns total missing values from each columns
raw_data.isnull().sum()
````

````python
# Handling the missing values
````

````python
#Display missing values percentage
raw_data.isnull().sum() / raw_data.shape[0] * 100
````

````python
#drop columns by using column names
raw_data.drop(['symboling', 'normalized-losses'], axis=1, inplace = True)
````

````python
raw_data['price'].dtype
````

````python
raw_data['price'] = raw_data['price'].astype('float64')
````

````python
raw_data['price'].dtype
````

````python
vehicle_df = raw_data.copy()
````

````python
vehicle_df
````

````python
#median
vehicle_df['price'].median()
````

````python
vehicle_df.isnull().sum()
````

#### 13. Merge

````python
df_series1 = pd.Series({
    "United States": 12313,
    "Canada": 45700,
    "Germany": 46200,
    "United Kingdom": 46200,
    "Japan": 41400
}, name = 'GDP_per_captia')
````

````python
df_series2 = pd.Series({
    "United States": 1232343213,
    "Canada": 4572300,
    "Germany": 46232400,
    "United Kingdom": 412126200,
    "Japan": 4123400
}, name = 'population')
````

````python
df1 = pd.DataFrame(df_series1).reset_index()
````

````python
df1.columns = ['country','gpd_per_capita']
````

````python
df1
````

````python
df2 = pd.DataFrame(df_series2).reset_index()
````

````python
#rename existing column
df2.columns = ['country','population']
````

````python
df2
````

````python
#how : how to joing "inner, left, right, outer"
pd.merge(df1, df2, on='country', how='inner')
````
